## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
#Check number of unique items
unique_item_count = purchase_data['Item ID'].nunique()

#Find total revenue from sales
total_revenue = sum(purchase_data['Price'])
                                 
#Find total transactions
total_transactions = len(purchase_data['Price'])

#Find average sale price
average_price = total_revenue / total_transactions
```

```python
#Format data frame revenue and price columns to display as currency
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].astype(float)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map('${:,.2f}'.format)
             

#Format average price to display as currency
purchasing_analysis["Average Sale Price"] = purchasing_analysis["Average Sale Price"].astype(float)
purchasing_analysis["Average Sale Price"] = purchasing_analysis["Average Sale Price"].map('${:,.2f}'.format)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#This was a hard one! At first I went through and created bins like the above exercise and grouped the dataset by the age bracke bin
#Then I created separate data frames for each column of data to return for analysis
#Then merged all the data frames to display on the age bracket column
bracket = [0,9,14,19,24,29,34,39,np.inf]
bracket_labels = ["Under 10", "10-14", "15-19","20-24","25-29","30-34","35-39","Over 40"]
purchase_data['Age Bracket'] = pd.cut(purchase_data['Age'],bracket,labels=bracket_labels)
#create a new data frame that groups by age bracket
purchasing_by_age = purchase_data.groupby('Age Bracket')

#create a data frame binned by age for total purchase count
pba_total_purchases = pd.DataFrame(purchasing_by_age['Price'].count())

#create a data frame binned by age for average price
pba_average_purchase_price = pd.DataFrame(purchasing_by_age['Price'].mean())

#create a data frame binned by age for total price per person
pba_total_by_person = pd.DataFrame(purchasing_by_age['Price'].sum())

#create a data frame binned by age for average total per person
pba_avg_total_per_person = pd.DataFrame(purchasing_by_age['Price'].sum()/purchasing_by_age['SN'].nunique())

#merge the first two data frames into a purchasing by age final dataframe
pba_final = pd.merge(pba_total_purchases,pba_average_purchase_price, on='Age Bracket')

#merge the third data frame to the final version
pba_final = pd.merge(pba_final, pba_total_by_person, on='Age Bracket')

#merge the last data frame into the final version
pba_final = pd.merge(pba_final, pba_avg_total_per_person, on='Age Bracket')
#pba_final 
#checked the data frame, data looks correct now need to rename the column headers
pba_final = pba_final.rename(columns={'Price_x':'Total Purchases',
                                     'Price_y':'Average Purchase Price',
                                     'Price':'Total Price Per Person',
                                     '0':'Avg Total Purchase Per Person'})
#couldn't get the last column to rename with the column rename.  I don't know why
pba_final

```

```python
#I redid this section after reading an article Nolan turned me on to that had this elegant solution
# here's the stack overflow article: 
#https://stackoverflow.com/questions/44635626/rename-result-columns-from-pandas-aggregation-futurewarning-using-a-dict-with
#This passes the function to groupby as a dataframe while creating a series that will be the column headers
#this saves a ton of work making multiple lines of code for each df and then merging
#I used this method for the rest of the homework

#create function to create columns for each category and their calculations where the index will be the columns for the new dataframe
def purchase_analysis(x):
    purchase_analysis_columns = {'Total Purchases': x['Price'].count(),
        'Average Purchase Price': x['Price'].mean(),
        'Total Price Per Person': x['Price'].sum(),
        'Avg Total Purchase per Person': x['Price'].sum() / x['SN'].nunique()}
    return pd.Series(purchase_analysis_columns,index=['Total Purchases', 'Average Purchase Price','Total Price Per Person','Avg Total Purchase per Person'])
#create a new dataframe that sorts by age bin and uses the above funtion for the new columns and calculations
purchase_analysis_by_age = purchase_data.groupby(['Age Bracket']).apply(purchase_analysis)
#format total purchases column to be integers
purchase_analysis_by_age['Total Purchases'] = purchase_analysis_by_age['Total Purchases'].astype(int)
purchase_analysis_by_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create new function for top spenders that will create analysis columns for total purchases, average purchase price and total revenue
def top_spenders(x):
    top_spenders_columns = {'Total Purchases': x['Price'].count(),
        'Average Purchase Price': x['Price'].mean(),
        'Total Purchase Value': x['Price'].sum()}
    return pd.Series(top_spenders_columns, index=['Total Purchases', 'Average Purchase Price','Total Purchase Value'])
#group by screen name and apply the function
top_spenders = purchase_data.groupby(['SN']).apply(top_spenders)
#change total purchases format to be int                            
top_spenders['Total Purchases'] = top_spenders['Total Purchases'].astype(int)
#sort the column by total revenue by a player
top_spenders = top_spenders.sort_values(['Total Purchase Value'], ascending=False)
top_spenders.head()                            
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create a function to return purchase count, item price, and total purchase value
#had to add .max() for the item price because just leaving it as ['Price'] was returning weird values, .min() returned the same
def popular_item_analysis(x):
    item_columns = {"Purchase Count": x['Price'].count(),
                    "Item Price": x['Price'].max(),
                    "Total Purchase Value": x['Price'].sum()}
    return pd.Series(item_columns, index=['Purchase Count', 'Item Price','Total Purchase Value'])

#run the function against the data set to return the desired columns
popular_items = purchase_data.groupby(['Item ID','Item Name']).apply(popular_item_analysis)
#sort by most purchased
popular_items = popular_items.sort_values('Purchase Count', ascending=False)
#format purchase count to display as integer
popular_items['Purchase Count']=popular_items['Purchase Count'].astype(int)

popular_items.head()
```
